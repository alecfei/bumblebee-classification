```python
# Define the directory where you want to save the augmented images
save_dir = 'augmented_barbutellus'

# Check if the directory exists; if not, create it
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# Create an ImageDataGenerator for augmentation
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

# Path to the directory containing your 30 images
data_dir = pathlib.Path('/home/hduser/Desktop/augmented data/barbutellus')

# List all image files in the directory
image_files = list(data_dir.glob('*'))

# Set the number of augmented images to generate for each input image
num_augmentations = 10

# Iterate through the image files and apply augmentation
for image in image_files:
    img = load_img(image)  # Load the image
    x = img_to_array(img)  # Convert to a Numpy array
    x = x.reshape((1,) + x.shape)  # Reshape for augmentation

    # Apply augmentation and save the results
    i = 0
    for batch in datagen.flow(x, batch_size=1, save_to_dir=save_dir, save_prefix='augmented_barbutellus', save_format='jpg'):
        i += 1
        if i >= num_augmentations:  # Generate 10 augmented versions of each image
            break
```
