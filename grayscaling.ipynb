```python
file1 = '/home/hduser/Desktop/testing2/barbutellus/*.jpg' 
glob.glob(file1)
# Using List Comprehension to read all images
barbutellus = [cv2.imread(image) for image in glob.glob(file1)]
```

```python
# Define size of the figure
fig=plt.figure(figsize=(12, 12))

# Define row and cols in the figure
rows, cols = 5, 6

# Display all images
for i in range(0, len(barbutellus), rows*cols):
    for j in range(0, cols*rows):
        fig.add_subplot(rows, cols, j+1)
        plt.imshow(barbutellus[i+j])
    plt.show()
```

```python
# Create a directory to store the converted images
output_folder = "barbutellus gray"
os.makedirs(output_folder, exist_ok=True)
```

```python
barbutellus_gray = []

for image in barbutellus:
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    barbutellus_gray.append(image_gray)
    
    # Define the path to save the converted image in the output folder
    output_path = os.path.join(output_folder, f"barbutellus_gray_{len(barbutellus_gray)}.jpg")
    
    # Save the grayscale image
    cv2.imwrite(output_path, image_gray)
```
